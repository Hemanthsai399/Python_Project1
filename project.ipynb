EXPLORATORY DATA ANALYSIS

GENDER

```python
data1.columns
```

```python
gen=sns.countplot(x='Gender',data=data1)
```

```python
gen=sns.countplot(x='Gender',data=data1)
for bars in gen.containers:
    gen.bar_label(bars)
```

```python
data1.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
```

```python
sales_gen=data1.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.barplot(data=sales_gen,x='Gender',y='Amount');
```

AGE

```python
age=sns.countplot(data=data1,x='Age Group',hue='Gender')
```

```python
age=sns.countplot(data=data1,x='Age Group',hue='Gender')
for bars in age.containers:
    age.bar_label(bars);
```

```python
data1.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
```

```python
age_am=data1.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.barplot(data=age_am,x='Age Group',y='Amount');
```

STATE

```python
data1.columns
```

```python
data1.groupby(['State'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False)
```

```python
#total states by orders
sales_state=data1.groupby(['State'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False)
sns.set(rc={'figure.figsize':(16,6)})
sns.barplot(data=sales_state,x='State',y='Orders');
```

```python
#Total amount from states
state_am=data1.groupby(['State'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.barplot(data=state_am,x='State',y='Amount');
```
